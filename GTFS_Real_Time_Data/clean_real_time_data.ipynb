```python
# Libraries
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
# Constant Variables
# Filenames
EXTRACTED_REAL_TIME_DATA_FILENAME = 'extracted_real_time_data.json'
CLEANED_REAL_TIME_DATA_FILENAME = 'cleaned_real_time_data.json'
```

```python
# Read the data frame from a json file
df = pd.read_json(EXTRACTED_REAL_TIME_DATA_FILENAME, dtype={'RecordedDate': 'str'})
df
```

```python
# Convert the date and time strings to datetime
def convert_to_datetime(date, time):
    return datetime.strptime(date + time, '%Y%m%d%I:%M:%S %p')
df['RecordedDateTime'] = df.apply(lambda row: convert_to_datetime(row['RecordedDate'], row['RecordedTime']), axis=1)
df
```

```python
# Drop the unused columns
df.drop(['VehicleNo', 'TripId', 'Destination', 'Pattern', 'RecordedTime', 'RouteMap', 'RecordedDate'], axis=1, inplace=True)
df
```

```python
# Drop all null values
df.dropna(axis=1, inplace=True)
df
```

```python
# Drop all duplicates
df.drop_duplicates(inplace=True)
df
```

```python
# Filter the recorded hour bewteen 7:00 to 23:00
df = df[(df['RecordedDateTime'].dt.hour >= 7) & (df['RecordedDateTime'].dt.hour < 23)]
df
```

```python
# Filter the recorded date bewteen April 7 to April 9
df = df[(df['RecordedDateTime'].dt.day >= 7) & (df['RecordedDateTime'].dt.day <= 9)]
df
```

```python
# Get the unique values of RouteNo
df['RouteNo'].unique()
```

```python
# Get the unique values of Direction
df['Direction'].unique()
```

```python
# Get the stats of the data frame
df.describe()
```

```python
# Plot the graph of Latitude
plt.title('Boxplot of Latitude')
plt.boxplot(df['Latitude'])
```

```python
# Plot the graph of Longitude
plt.title('Boxplot of Longitude')
plt.boxplot(df['Longitude'])
```

```python
# Remove all outliers of Latitude
latitude_head = df['Latitude'].quantile(0.01)
latitude_tail = df['Latitude'].quantile(0.99)
df = df[(df['Latitude'] >= latitude_head) & (df['Latitude'] <= latitude_tail)]
df
```

```python
# Remove all outliers of Longitude
longitude_head = df['Longitude'].quantile(0.01)
longitude_tail = df['Longitude'].quantile(0.99)
df = df[(df['Longitude'] >= longitude_head) & (df['Longitude'] <= longitude_tail)]
df
```

```python
# Plot the graph of Latitude
plt.title('Boxplot of Latitude')
plt.boxplot(df['Latitude'])
```

```python
# Plot the graph of Longitude
plt.title('Boxplot of Longitude')
plt.boxplot(df['Longitude'])
```

```python
# Save the cleaned real time data frame
df.to_json(CLEANED_REAL_TIME_DATA_FILENAME, orient='records')
```
