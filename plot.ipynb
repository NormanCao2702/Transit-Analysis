```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame setup
data = {
    'Period': ['Morning Peak', 'After-noon Peak', 'Evening Peak', 'Late-Night'],
    'Monday': [10, 5, 15, 20],
    'Tuesday': [12, 6, 14, 19],
    # Add the rest of the days and their data
}

df = pd.DataFrame(data).set_index('Period')

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df, annot=True, cmap='coolwarm', fmt="g")
plt.title('Average Wait Times for Buses to SFU')
plt.xlabel('Day of the Week')
plt.ylabel('Period')
plt.show()
```
